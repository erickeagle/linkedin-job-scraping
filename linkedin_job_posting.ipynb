```python
a=soup.findAll('div', attrs={'class':"base-card base-card--link base-search-card base-search-card--link job-search-card"})
```

```python
len(a)
```

```python
links=[]
for i in a:
    links.append(i.find('a', attrs={'class':"base-card__full-link"})['href'])
```

```python
len(links)
```

```python
jobUrl=[]
datePosted=[]
jobTitle=[]
skills=[]
industry=[]
industryAddressCountry=[]
industryAddressLocality=[]
industryAddressRegion=[]
industryStreetAddress=[]
employmentType=[]
hiringOrganizationName=[]
hiringOrganizationUrl=[]
jobDescription=[]


for utl in tqdm(range((len(links)))):
    #print(links[utl])
    try:
        driver = webdriver.Firefox(executable_path=r'geckodriver.exe')
        driver.wait = WebDriverWait(driver, 5)
        driver.maximize_window()
        driver.get(links[utl])
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(7)
        height = driver.execute_script("return document.documentElement.scrollHeight")
        lastheight = 0
        while True:
            if lastheight == height:
                break
            lastheight = height
            driver.execute_script("window.scrollTo(0, " + str(height) + ");")
            time.sleep(5)
            height = driver.execute_script("return document.documentElement.scrollHeight")
        soup = BeautifulSoup(driver.page_source, "html")
        
        jobUrl1=links[utl]
        datePosted1=json.loads(soup.find('script', type='application/ld+json').contents[0])['datePosted'] 
        jobTitle1=json.loads(soup.find('script', type='application/ld+json').contents[0])['title'] 
        skills1=json.loads(soup.find('script', type='application/ld+json').contents[0])['skills'] 
        industry1=json.loads(soup.find('script', type='application/ld+json').contents[0])['industry'] 
        industryAddressCountry1=json.loads(soup.find('script', type='application/ld+json').contents[0])['jobLocation']['address']['addressCountry']
        industryAddressLocality1=json.loads(soup.find('script', type='application/ld+json').contents[0])['jobLocation']['address']['addressLocality']
        industryAddressRegion1=json.loads(soup.find('script', type='application/ld+json').contents[0])['jobLocation']['address']['addressRegion']
        industryStreetAddress1=json.loads(soup.find('script', type='application/ld+json').contents[0])['jobLocation']['address']['streetAddress']
        employmentType1=json.loads(soup.find('script', type='application/ld+json').contents[0])['employmentType']
        hiringOrganizationName1=json.loads(soup.find('script', type='application/ld+json').contents[0])['hiringOrganization']['name']
        hiringOrganizationUrl1=json.loads(soup.find('script', type='application/ld+json').contents[0])['hiringOrganization']['sameAs']
        jobDescription1=json.loads(soup.find('script', type='application/ld+json').contents[0])['description']
        
        jobUrl.append(jobUrl1)
        datePosted.append(datePosted1)
        jobTitle.append(jobTitle1)
        skills.append(skills1)
        
        industry.append(industry1)
        industryAddressCountry.append(industryAddressCountry1)
        industryAddressLocality.append(industryAddressLocality1)
        industryAddressRegion.append(industryAddressRegion1)
        industryStreetAddress.append(industryStreetAddress1)
        employmentType.append(employmentType1)
        hiringOrganizationName.append(hiringOrganizationName1)
        hiringOrganizationUrl.append(hiringOrganizationUrl1)
        jobDescription.append(jobDescription1)
        
        
        
        driver.close()
    except:
        pass
```

```python
df = pd.DataFrame([jobUrl,
datePosted,
jobTitle,
skills,
industry,
industryAddressCountry,
industryAddressLocality,
industryAddressRegion,
industryStreetAddress,
employmentType,
hiringOrganizationName,
hiringOrganizationUrl,
jobDescription]).transpose()


df.columns = ['jobUrl',
'datePosted',
'jobTitle',
'skills',
'industry',
'industryAddressCountry',
'industryAddressLocality',
'industryAddressRegion',
'industryStreetAddress',
'employmentType',
'hiringOrganizationName',
'hiringOrganizationUrl',
'jobDescription']
```
